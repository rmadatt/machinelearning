```python
from sentence_transformers import SentenceTransformer

# Load a pre-trained transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Poetic testimony fragments
fragments = [
    "I chose nourishment over absence.",
    "I refused to be erased.",
    "Joy is a quiet rebellion.",
    "Grief is a sovereign archive.",
    "I forgave but never forgot."
]
```

```python
# Convert fragments to vector embeddings
embeddings = model.encode(fragments)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Compute similarity matrix
similarity_matrix = cosine_similarity(embeddings)

# Display as DataFrame
df = pd.DataFrame(similarity_matrix, index=fragments, columns=fragments)
df.style.background_gradient(cmap='coolwarm')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(similarity_matrix, xticklabels=fragments, yticklabels=fragments, annot=True, cmap='coolwarm')
plt.title("Semantic Similarity Between Fragments")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
